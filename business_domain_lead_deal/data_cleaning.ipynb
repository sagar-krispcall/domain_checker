```python
import pandas as pd
import numpy as np
```

```python
deal1 = pd.read_csv('deals-19264736-555.csv')
deal2 = pd.read_csv('deals-19264736-556.csv')
bd = pd.read_csv('busines_domains_ac.csv')
```

```python
deal1.head(2)
```

```python
deal1["email"] = deal1["Person - Email - Other"].fillna(deal1["Person - Email - Work"])
deal1 = deal1.rename(columns={'Person - Country': 'Country'})
deal1 = deal1[['email', 'Country']]
print(len(deal1))
deal1.head(2)
```

```python
deal1 = deal1.copy()  

deal1['email'] = deal1['email'].str.split(r'[;, ]+')
deal1 = deal1.explode('email').reset_index(drop=True)
deal1['email'] = deal1['email'].str.strip()
deal1 = deal1[deal1['email'].notna() & (deal1['email'] != '')]
deal1 = deal1.sort_values(by='email', ascending=False)

print(len(deal1))
deal1.head(2)
```

```python
deal1 = deal1.drop_duplicates(subset = 'email')
print(len(deal1))
deal1.head(2)
```

```python
deal2.head(2)
```

```python
deal2 = deal2.rename(columns={
    'Deal - Title': 'email',
    'Person - Country': 'country'
})
deal2 = deal2[['email', 'country']]
print(len(deal2))
deal2.head(2)
```

```python
deal2 = deal2.drop_duplicates(subset = 'email')
print(len(deal2))
deal2.head(2)
```

```python
bd.head(2)
```

```python
bd['email'] = np.where(
    bd['$distinct_id_before_identity'].str.contains('@', na=False),
    bd['$distinct_id_before_identity'],
    np.where(
        bd['distinct_id'].str.contains('@', na=False),
        bd['distinct_id'],
        np.nan  # set NaN if no '@' in either
    )
)
print(len(bd))
bd.head(2)
```

```python
bd = bd[['email']]
bd = bd.drop_duplicates()
print(len(bd))
bd.head(2)
```

```python
deal_combined = pd.merge(deal1, deal2, on = 'email', how = 'outer')
print(len(deal_combined))
deal_bd_combined = pd.merge(deal_combined, bd, on = 'email', how = 'outer')
print(len(deal_bd_combined))
deal_bd_combined.head(2)
```

```python
deal_combined['country'] = deal_combined['country'].fillna(deal_combined['Country'])
deal_combined = deal_combined.drop(columns=['Country'])
deal_combined.head(10)
```

```python
deal_combined['email'] = deal_combined['email'].str.lower().str.strip()
deal_combined = deal_combined[~deal_combined['email'].str.contains(' ', na=False)]
deal_combined = deal_combined.reset_index(drop=True)
print(len(deal_combined))
```

```python
deal_combined[['username', 'domain']] = deal_combined['email'].str.split('@', expand=True)
deal_combined = deal_combined.drop_duplicates(subset='domain')
deal_combined = deal_combined[['domain', 'country']]
deal_combined
```

```python
deal_combined.to_csv('deal_data.csv', index = False)
```
