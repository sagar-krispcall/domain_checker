```python
import pandas as pd
```

```python
df = pd.read_csv('Enterprise Clients.csv')
print(df.shape)
df.head(2)
```

```python
print(df['email'].nunique())
```

```python
df_active = df[df['subscription_status'] == 'Active']
print(len(df_active))
```

```python
df_active['Pay'].dtypes
```

```python
df_active_pay = df_active[df_active['Pay']>900]
print(len(df_active_pay))
```

```python
final_df = df_active_pay.copy()
final_df['email'] = final_df['email'].str.lower().str.strip()
final_df = final_df[~final_df['email'].str.contains(' ', na=False)]
final_df = final_df.reset_index(drop=True)
print(len(final_df))
```

```python
final_df[['username', 'domain']] = final_df['email'].str.split('@', expand=True)
final_df = final_df.drop_duplicates(subset='domain')
final_df = final_df[['domain']]
final_df
```

```python
final_df.to_csv('enterprise_domain(active_pay_900).csv', index = False)
```
