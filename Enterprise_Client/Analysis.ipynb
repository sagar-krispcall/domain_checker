```python
import pandas as pd
```

```python
df = pd.read_csv('Enterprise Clients.csv')
df2 = pd.read_csv('enterprise_domain(active_pay_400_550)_enriched.csv')
print(df.shape)
df.head(2)
```

```python
print(df['email'].nunique())
```

```python
df_active = df[df['subscription_status'] == 'Active']
print(len(df_active))
```

```python
df_active['Pay'].dtypes
```

```python
df_active_pay = df_active[df_active['Pay'].between(400,550)]
print(len(df_active_pay))
```

```python
df_active_pay = df_active_pay.copy()
df_active_pay['email'] = df_active_pay['email'].str.lower().str.strip()
df_active_pay = df_active_pay[~df_active_pay['email'].str.contains(' ', na=False)]
df_active_pay = df_active_pay.reset_index(drop=True)
print(len(df_active_pay))
```

```python
df_active_pay[['username', 'domain']] = df_active_pay['email'].str.split('@', expand=True)
#df_active_pay = df_active_pay.drop_duplicates(subset='domain')
df_active_pay.head(2)
```

```python
final_df = df_active_pay[['email', 'domain']]
final_df
```

```python
final_df.to_csv('enterprise_domain(active_pay_400_550).csv', index = False)
```

```python
final = pd.merge(df_active_pay, df2, on = ['domain','email'], how = 'left')
print(len(final))
final.head(5)
```

```python
final = final.to_csv('active_pay_400_550_domain_classify.csv', index = False)
```
